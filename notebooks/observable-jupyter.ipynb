# USGS earthquakes

[GeoJSON feed](https://earthquake.usgs.gov/earthquakes/feed/v1.0/geojson.php)





```python
# Get some GeoJSON earthquakes
import requests

url_month = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.geojson"
url_week = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson"
response = requests.get(url=url_month)
quakes = response.json()
print("There are {} earthquakes in the USGS response".format(len(quakes['features'])))
```

```python
# Serialize the data into a JSON string
import json

# First, filter the quakes, just for the heck of it.
filtered_quakes = [d for d in quakes['features'] if d['geometry']['coordinates'][0] > 0]
print("There are {} filtered quakes in the GeoJSON data string\n".format(len(filtered_quakes)))
quakes['features'] = filtered_quakes

# Create the GeoJSON string
geojson_string = json.dumps(quakes)
```

```python
embed('@pbogden/brushable-earthquakes-iii', cells=['mymap', 'mychart', 'map_filter', 'myreport'], inputs={"source": geojson_string})
```
